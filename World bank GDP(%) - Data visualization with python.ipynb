```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df_gdp = pd.read_excel('world_gdp.xls')
df_gdp.head()
```

```python
df_gdp = df_gdp.fillna(0)
df_gdp.head(2)
```

```python
years = list(map(str, range(1960, 2021)))
df_gdp.set_index('Country Name', inplace=True)
```

```python
fig = plt.figure(figsize=(13,4))
df_gdp.loc['India', years].plot(kind='bar')
plt.title('GDP growth rate of India from 1960 to 2020')
plt.xlabel('Year')
plt.ylabel('GDP growth rate (in %)')
plt.show()
```

```python
years_60s = list(map(str, range(1960, 1970)))
years_70s = list(map(str, range(1970, 1980)))
years_80s = list(map(str, range(1980, 1990)))
years_90s = list(map(str, range(1990, 2000)))
years_2000s = list(map(str, range(2000, 2010)))
years_2010s = list(map(str, range(2010, 2020)))
years_2020s = list(map(str, range(2020, 2021)))

df_60s = df_gdp.loc[:, years_60s].mean(axis=1)
df_70s = df_gdp.loc[:, years_70s].mean(axis=1)
df_80s = df_gdp.loc[:, years_80s].mean(axis=1)
df_90s = df_gdp.loc[:, years_90s].mean(axis=1)
df_2000s = df_gdp.loc[:, years_2000s].mean(axis=1)
df_2010s = df_gdp.loc[:, years_2010s].mean(axis=1)
df_2020s = df_gdp.loc[:, years_2020s].mean(axis=1)

df_newGdp = pd.DataFrame({'1960': df_60s, '1970': df_70s, '1980': df_80s, '1990': df_90s, '2000': df_2000s, '2010': df_2010s, '2020': df_2020s})
df_newGdp.head()
```

```python
years = ['1960', '1970', '1980', '1990', '2000', '2010', '2020']
country = ['India']
India = df_newGdp.loc[country, years]
India
```

```python
years = ['1960', '1970', '1980', '1990', '2000', '2010', '2020']

fig = plt.figure(figsize=(3,3))
df_newGdp.loc['India', years].plot(kind='bar')
plt.title('Average GDP growth rate of India from decades 60s to 2020s')
plt.xlabel('Decade')
plt.ylabel('GDP growth rate (in %)')
plt.show()
```

```python
years = list(map(str, range(1990, 2021)))

fig = plt.figure(figsize=(10,3))
df_gdp.loc['India', years].plot(kind='line',alpha=1, color='#a800b1', linestyle='-', linewidth=0.5, marker='.', markeredgecolor='red', 
         markeredgewidth=1, markersize=10, markerfacecolor='yellow', visible=True)
plt.title('GDP growth rate of India in last 20 years [1990-2020]')
plt.xlabel('Year')
plt.ylabel('GDP growth rate (in %)')
plt.show()
```

```python
countries = ['United States','China','Japan','Germany','United Kingdom','France','India','Italy','Brazil','Canada']


fig = plt.figure(figsize=(10,4))
df_gdp.loc[countries, '2020'].plot(kind='line',alpha=1, color='#a800b1', linestyle='-', linewidth=0.5, marker='.', markeredgecolor='red', 
         markeredgewidth=1, markersize=10, markerfacecolor='yellow', visible=True)
plt.title('Average GDP growth rate of top ten economic superpowers in yr 2020')
plt.xlabel('Countries')
plt.ylabel('GDP growh rate (in %)')
plt.show()
```

```python
countries = ['United States','China','Japan','Germany','United Kingdom','France','India','Italy','Brazil','Canada']


fig = plt.figure(figsize=(5,4))
df_gdp.loc[countries, '2020'].plot(kind='barh',color='red')
df_gdp.loc[countries, '2019'].plot(kind='barh',color='green')

plt.title('Average GDP growth rate of top ten economic superpowers in yr 2019 & 2020')
plt.xlabel('Countries')
plt.ylabel('GDP growh rate (in %)')
plt.legend()
plt.show()
```

```python
income_gdp = df_gdp.groupby('IncomeGroup')['2020'].agg('mean')
print(income_gdp)
```

```python
region_gdp = df_gdp.groupby('Region')['2020'].agg('mean')
print(region_gdp)
```

```python
fig = plt.figure()

ax0 = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)

income_gdp.plot(kind='bar',color='red',figsize=(15,4),ax=ax0)
ax0.set_title('Average GDP growth rate of income group wise in yr 2020')
ax0.set_xlabel('Income group')
ax0.set_ylabel('GDP growth rate (in %)')

region_gdp.plot(kind='bar',color='red', figsize=(15, 4), ax=ax1)
ax1.set_title ('Average GDP growth rate of continent wise in yr 2020')
ax1.set_ylabel('GDP growth rate (in %)')
ax1.set_xlabel('Continent')

plt.show()
```

```python
years = list(map(str, range(1980, 2021)))

df_CI= df_gdp.loc[['United States','China','Japan','Germany','United Kingdom','France','India','Italy','Brazil','Canada'], years].transpose()
df_CI.head()
```

```python
fig = plt.figure()

ax0 = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)

df_CI.plot(kind='box',color='blue',vert=False,figsize=(20,6),ax=ax0)
ax0.set_title('Box plot of top ten economic superpowers (1980-2021)')
ax0.set_xlabel('GDP growth rate')
ax0.set_ylabel('Countries')

df_CI.plot(kind='line', figsize=(20, 6), ax=ax1)
ax1.set_title ('Line Plots of top ten economic superpowers (1980-2021)')
ax1.set_ylabel('GDP growth rate')
ax1.set_xlabel('Years')

plt.show()
```

```python
df_gdp.sort_values(by='2020', ascending=False, axis=0, inplace=True)

df_top5 = df_gdp.head(5)
df_top5 = df_top5['2020'].transpose()

print(df_top5)

df_top5.plot(kind='line', figsize=(10, 3), alpha=1, color='#a800b1', linestyle='-', linewidth=0.5, marker='.', markeredgecolor='red', 
         markeredgewidth=1, markersize=10, markerfacecolor='yellow', visible=True)
plt.title('Top five countries with highest GDP growth rate in 2020')
plt.xlabel('Countries')
plt.ylabel('GDP growth rate (in %)')
plt.show()
```

```python
df_gdp.sort_values(by='2020', ascending=False, axis=0, inplace=True)

df_low5 = df_gdp.tail(5)
df_low5 = df_low5['2020'].transpose()

print(df_low5)

df_low5.plot(kind='line', figsize=(10, 3), alpha=1, color='#a800b1', linestyle='-', linewidth=0.5, marker='.', markeredgecolor='red', 
         markeredgewidth=1, markersize=10, markerfacecolor='yellow', visible=True)
plt.title('Top five countries with lowest GDP growth rate in 2020')
plt.xlabel('Countries')
plt.ylabel('GDP growth rate (in %)')
plt.show()
```
