```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_me = pd.read_excel('military_expenditure.xls')
df_me = df_me.fillna(0)

years = list(map(str, range(1980, 2021)))
df_me.set_index('Country', inplace=True)

df_me.head(2)
```

```python
fig = plt.figure(figsize=(13,4))

df_me.loc['India', years].plot(kind='bar',color='orange')
df_me.loc['China', years].plot(kind='bar',color='green')

plt.title('Military expenditure of India & China (% of GDP) [1980-2020]')
plt.xlabel('Year')
plt.ylabel('Military expenditure (% of GDP)')
plt.show()
```

```python
df = df_me.loc[['India','China','United States'],years].transpose()
df.head()
```

```python
df.plot(kind='hist', figsize=(10,6),
       bins=10,
       alpha=1,
       color=['coral', 'darkslateblue', 'mediumseagreen'])

plt.title('Histogram of military expenditure of India, China & United States (% of GDP) [1980-2020]')
plt.ylabel('Number of Years')
plt.xlabel('Military expenditure (% of GDP)')

plt.show()
```

```python
continent = df_me.groupby('Continent', axis=0).mean()
continent
```

```python
income_gp = df_me.groupby('Income_group', axis=0).mean()
income_gp
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink','yellow']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1,0] 

continent['2020'].plot(kind='pie',
                            figsize=(15,6),
                            autopct='%1.1f%%',
                            startangle=90,
                            shadow=True,
                            labels=None,
                            pctdistance=1.12,
                            colors=colors_list,
                            explode=explode_list
                            )

plt.title('Military expenditure continent wise (% of GDP) in year-2020', y=1.12)
plt.axis('equal')

plt.legend(labels=continent.index, loc='upper left')

plt.show()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode_list = [0.1, 0.1, 0, 0] 

income_gp['2020'].plot(kind='pie',
                            figsize=(15,6),
                            autopct='%1.1f%%',
                            startangle=90,
                            shadow=True,
                            labels=None,
                            pctdistance=1.12,
                            colors=colors_list,
                            explode=explode_list
                            )

plt.title('Military expenditure income group wise (% of GDP) in year-2020', y=1.12)
plt.axis('equal')

plt.legend(labels=income_gp.index, loc='upper left')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=[10,5])
sns_hist = sns.countplot(y='Continent',hue='Income_group',data=df_me,dodge=False,orient='h')

plt.title('Count of countries continent & income group wise')
```

```python
fig, ax = plt.subplots(figsize = [15,5])

ax = sns.boxplot(x="Continent", y="2020", 
                 hue="Income_group",
                 data=df_me, 
                 palette="Set3"
                )

plt.title('Box plot of military expenditure continent & Income group wise (% of GDP) in year-2020')
```

```python
df_me = pd.read_excel('military_expenditure.xls')
df_me.sort_values(by='2020', ascending=False, axis=0, inplace=True)

me_top10 = df_me.head(10)
me_top10
```

```python
fig = plt.figure(figsize=(15,5))

x=me_top10['Country']
y=me_top10['2020']
plt.title('Top ten countries with highest military expenditure (% of GDP) in year-2020')
plt.xlabel('Countries')
plt.ylabel('Military expenditure (% of GDP)')
plt.plot(x, y, alpha=1, color='red', linestyle='-', linewidth=0.5, marker='.', markeredgecolor='red', 
         markeredgewidth=1, markersize=10, markerfacecolor='yellow', visible=True)
plt.show()
```

```python
#military expenditure (% of GDP) of above countries in last five years [2016-2020]

fig = plt.figure(figsize=(8,3))

plt.plot(me_top10['Country'],me_top10['2016'],label='Militry expences in Yr-2016',marker='.')
plt.plot(me_top10['Country'],me_top10['2017'],label='Militry expences in Yr-2017',marker='.')
plt.plot(me_top10['Country'],me_top10['2018'],label='Militry expences in Yr-2018',marker='.')
plt.plot(me_top10['Country'],me_top10['2019'],label='Militry expences in Yr-2019',marker='.')
plt.plot(me_top10['Country'],me_top10['2020'],label='Militry expences in Yr-2020',marker='.')

plt.title('Military expenditure (% of GDP) in last five years [2016-2020]')
plt.xlabel('Country')
plt.ylabel('military expenditure (% of GDP)')
plt.legend()
plt.show()
```
